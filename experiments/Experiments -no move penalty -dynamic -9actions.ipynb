```python
from qlearning import run_experiment
from env import layout_original, layout_original_walled, layout_nowalls
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Expirments no move penalty, dynamic env, 5 actions

```python
SIZE = 11
LAYOUT = layout_original_walled
SHOW = False
EPISODES = 5000
N_ACTIONS = 9
DYNAMIC = True
DISCOUNT = 0.95
MOVE_PENALTY = 0
TARGET_REWARD = 25
WALL_PENALTY = -5
RS_PENALTY = -1
SHOW_EVERY = 100
EPSILON_START=0.3
EPSILON_END=0.01
LEARNING_RATE_START = 0.2
LEARNING_RATE_END = 0.05
SAVE_Q_TABLE = False
SAVE_RESULTS = True

n_experiments = 10
```

```python
DF_Unshielded_QL = pd.DataFrame()
for i in range(n_experiments):
    Unshielded_QL = run_experiment(SHIELD_ON = False,
                                   CQL = False,
                                   RS = False,
                                   start_q_table = None,
                                   SIZE = SIZE, LAYOUT = LAYOUT, SHOW = SHOW, EPISODES = EPISODES,
                                   N_ACTIONS = N_ACTIONS, DYNAMIC=DYNAMIC, DISCOUNT = DISCOUNT,
                                   MOVE_PENALTY = MOVE_PENALTY, TARGET_REWARD = TARGET_REWARD, 
                                   WALL_PENALTY = WALL_PENALTY, RS_PENALTY = RS_PENALTY, SHOW_EVERY = SHOW_EVERY,
                                   EPSILON_START=EPSILON_START, EPSILON_END=EPSILON_END,
                                   LEARNING_RATE_START = LEARNING_RATE_START, LEARNING_RATE_END = LEARNING_RATE_END,
                                   SAVE_Q_TABLE = SAVE_Q_TABLE, SAVE_RESULTS = SAVE_RESULTS)
    tmp_df = pd.DataFrame(Unshielded_QL)
    DF_Unshielded_QL = DF_Unshielded_QL.append(tmp_df)

DF_Unshielded_QL.reset_index(inplace=True)
```

```python
DF_Shielded_QL = pd.DataFrame()
for i in range(n_experiments):
    Shielded_QL = run_experiment(SHIELD_ON = True,
                                   CQL = False,
                                   RS = False,
                                   start_q_table = None,
                                   SIZE = SIZE, LAYOUT = LAYOUT, SHOW = SHOW, EPISODES = EPISODES,
                                   N_ACTIONS = N_ACTIONS, DYNAMIC=DYNAMIC, DISCOUNT = DISCOUNT,
                                   MOVE_PENALTY = MOVE_PENALTY, TARGET_REWARD = TARGET_REWARD, 
                                   WALL_PENALTY = WALL_PENALTY, RS_PENALTY = RS_PENALTY, SHOW_EVERY = SHOW_EVERY,
                                   EPSILON_START=EPSILON_START, EPSILON_END=EPSILON_END,
                                   LEARNING_RATE_START = LEARNING_RATE_START, LEARNING_RATE_END = LEARNING_RATE_END,
                                   SAVE_Q_TABLE = SAVE_Q_TABLE, SAVE_RESULTS = SAVE_RESULTS)
    tmp_df = pd.DataFrame(Shielded_QL)
    DF_Shielded_QL = DF_Shielded_QL.append(tmp_df)

DF_Shielded_QL.reset_index(inplace=True)
```

```python
DF_Shielded_CQL = pd.DataFrame()
for i in range(n_experiments):
    Shielded_CQL = run_experiment(SHIELD_ON = True,
                                   CQL = True,
                                   RS = True,
                                   start_q_table = None,
                                   SIZE = SIZE, LAYOUT = LAYOUT, SHOW = SHOW, EPISODES = EPISODES,
                                   N_ACTIONS = N_ACTIONS, DYNAMIC=DYNAMIC, DISCOUNT = DISCOUNT,
                                   MOVE_PENALTY = MOVE_PENALTY, TARGET_REWARD = TARGET_REWARD, 
                                   WALL_PENALTY = WALL_PENALTY, RS_PENALTY = RS_PENALTY, SHOW_EVERY = SHOW_EVERY,
                                   EPSILON_START=EPSILON_START, EPSILON_END=EPSILON_END,
                                   LEARNING_RATE_START = LEARNING_RATE_START, LEARNING_RATE_END = LEARNING_RATE_END,
                                   SAVE_Q_TABLE = SAVE_Q_TABLE, SAVE_RESULTS = SAVE_RESULTS)
    tmp_df = pd.DataFrame(Shielded_CQL)
    DF_Shielded_CQL = DF_Shielded_CQL.append(tmp_df)

DF_Shielded_CQL.reset_index(inplace=True)
```

```python
DF_Shielded_QL_RS = pd.DataFrame()
for i in range(n_experiments):
    Shielded_QL_RS = run_experiment(SHIELD_ON = True,
                                   CQL = False,
                                   RS = True,
                                   start_q_table = None,
                                   SIZE = SIZE, LAYOUT = LAYOUT, SHOW = SHOW, EPISODES = EPISODES,
                                   N_ACTIONS = N_ACTIONS, DYNAMIC=DYNAMIC, DISCOUNT = DISCOUNT,
                                   MOVE_PENALTY = MOVE_PENALTY, TARGET_REWARD = TARGET_REWARD, 
                                   WALL_PENALTY = WALL_PENALTY, RS_PENALTY = RS_PENALTY, SHOW_EVERY = SHOW_EVERY,
                                   EPSILON_START=EPSILON_START, EPSILON_END=EPSILON_END,
                                   LEARNING_RATE_START = LEARNING_RATE_START, LEARNING_RATE_END = LEARNING_RATE_END,
                                   SAVE_Q_TABLE = SAVE_Q_TABLE, SAVE_RESULTS = SAVE_RESULTS)
    tmp_df = pd.DataFrame(Shielded_QL_RS)
    DF_Shielded_QL_RS = DF_Shielded_QL_RS.append(tmp_df)

DF_Shielded_QL_RS.reset_index(inplace=True)
```

## Plot together

```python
agents = [(DF_Unshielded_QL[DF_Unshielded_QL["index"] != 0], "Unshielded_QL"),
         (DF_Shielded_QL[DF_Shielded_QL["index"]!=0], "Shielded_QL"),
         (DF_Shielded_QL_RS[DF_Shielded_QL_RS["index"]!=0], "Shielded_QL_RS"),
         (DF_Shielded_CQL[DF_Shielded_CQL["index"]!=0], "Shielded_CQL")]

sns.set_theme(style="darkgrid")



for agent_data in agents:
    ax = sns.lineplot(data=agent_data[0], x="index", y=0, err_style="band", legend="brief",label=f"{agent_data[1]}")

    
ax.set(xlabel='Episodes', ylabel=f"Rewards {SHOW_EVERY} episodes moving avg", title='')
plt.show()
```
