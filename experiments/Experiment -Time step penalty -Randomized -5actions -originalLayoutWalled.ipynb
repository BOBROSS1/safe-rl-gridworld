```python
from qlearning import run_experiment, generate_df
from env import layout_original, layout_original_walled, layout_nowalls, layout_original_walled_big
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Expirments no time step penalty, Randomized env, 5 actions, OriginalLayoutWalled

```python
n_experiments = 30
SIZE = 11
LAYOUT = layout_original_walled
# SHOW = False
EPISODES = 150
N_ACTIONS = 5 # N_ACTIONS must be 5 or 9
FIXED = False
# DISCOUNT = 0.95
MOVE_PENALTY = -1
# TARGET_REWARD = 25
# WALL_PENALTY = -5
# RS_PENALTY = -1
SHOW_EVERY = 10
# EPSILON_START=0.3
# EPSILON_END=0.01
# LEARNING_RATE_START = 0.2
# LEARNING_RATE_END = 0.05
```

## 4 types of agents
1. TRADITIONAL_QL
2. TRADITIONAL_QL_SHIELDED
3. TRADITIONAL_QL_SHIELDED_RS
4. CONSTRAINED_QL_SHIELDED

```python
TRADITIONAL_QL_UNSHIELDED = run_experiment(n_experiments=n_experiments, agent_type="TQLU", N_ACTIONS=N_ACTIONS,
                                           FIXED=FIXED, EPISODES=EPISODES, SHOW_EVERY=SHOW_EVERY,
                                           MOVE_PENALTY=MOVE_PENALTY, LAYOUT=LAYOUT, SIZE=SIZE)
DF_TRADITIONAL_QL_UNSHIELDED = generate_df(TRADITIONAL_QL_UNSHIELDED, plot=True)
```

```python
TRADITIONAL_QL_SHIELDED = run_experiment(n_experiments=n_experiments, agent_type="TQLS", N_ACTIONS=N_ACTIONS,
                                         FIXED=FIXED, EPISODES=EPISODES, SHOW_EVERY=SHOW_EVERY,
                                         MOVE_PENALTY=MOVE_PENALTY, LAYOUT=LAYOUT, SIZE=SIZE)
DF_TRADITIONAL_QL_SHIELDED = generate_df(TRADITIONAL_QL_SHIELDED, plot=True)
```

```python
TRADITIONAL_QL_SHIELDED_RS = run_experiment(n_experiments=n_experiments, agent_type="TQLSRS",
                                            FIXED=FIXED, N_ACTIONS=N_ACTIONS, EPISODES=EPISODES,
                                            SHOW_EVERY=SHOW_EVERY, MOVE_PENALTY=MOVE_PENALTY, LAYOUT=LAYOUT,
                                            SIZE=SIZE)
DF_TRADITIONAL_QL_SHIELDED_RS = generate_df(TRADITIONAL_QL_SHIELDED_RS, plot=True)
```

```python
CQL_SHIELDED = run_experiment(n_experiments=n_experiments, agent_type="CQLS", EPISODES=EPISODES,
                              N_ACTIONS=N_ACTIONS, FIXED=FIXED, SHOW_EVERY=SHOW_EVERY,
                              MOVE_PENALTY=MOVE_PENALTY, LAYOUT=LAYOUT, SIZE=SIZE)
DF_CQL_SHIELDED = generate_df(CQL_SHIELDED, plot=True)
```

## Plot together

```python
agents = [(DF_TRADITIONAL_QL_UNSHIELDED, "TRADITIONAL_QL_UNSHIELDED (TQLU)"),
         (DF_TRADITIONAL_QL_SHIELDED, "TRADITIONAL_QL_SHIELDED (TQLS)"),
         (DF_TRADITIONAL_QL_SHIELDED_RS, "TRADITIONAL_QL_SHIELDED_RS (TQLSRS)"),
         (DF_CQL_SHIELDED, "CONSTRAINED_QL_SHIELDED (CQLS)")]

sns.set_theme(style="darkgrid")



for agent_data in agents:
    ax = sns.lineplot(data=agent_data[0], x="index", y=0, err_style="band", legend="brief",label=f"{agent_data[1]}")

    
ax.set(xlabel='Episodes', ylabel=f"Reward", title='')
plt.show()
```
